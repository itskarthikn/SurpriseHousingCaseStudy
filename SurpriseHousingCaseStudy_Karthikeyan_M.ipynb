#### Missing Value Treatment

```python
#check null values
null_cols = house.columns[house.isnull().any()]
null_house = house[null_cols].isnull().sum().to_frame(name='Null Count')\
          .merge(house[null_cols].isnull().mean().mul(100).to_frame(name='Null Percent'), left_index=True, right_index=True)
null_house_sorted = null_house.sort_values(by='Null Count', ascending=False)
print(null_house_sorted)
```

```python
#removing variables whose null percent is > 40%
house.drop(['PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu'] , axis = 1 , inplace = True)
house.dropna(subset=['Electrical'], inplace=True)
```
