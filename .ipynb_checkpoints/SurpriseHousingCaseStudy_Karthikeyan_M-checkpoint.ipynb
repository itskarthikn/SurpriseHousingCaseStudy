#### Missing Value Treatment

```python
#check null values
null_cols = house.columns[house.isnull().any()]
null_house = house[null_cols].isnull().sum().to_frame(name='Null Count')\
          .merge(house[null_cols].isnull().mean().mul(100).to_frame(name='Null Percent'), left_index=True, right_index=True)
null_house_sorted = null_house.sort_values(by='Null Count', ascending=False)
print(null_house_sorted)
```

```python
#removing variables whose null percent is > 40%
house=house.drop(['Id','PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu'] , axis = 1)
house=house.dropna(subset=['Electrical'], axis = 0)
```

```python
house[['LotFrontage', 'MasVnrArea','GarageYrBlt']] = house[['LotFrontage', 'MasVnrArea','GarageYrBlt']].fillna(house[['LotFrontage', 'MasVnrArea','GarageYrBlt']].median())
```

```python
house['GarageType'] = house['GarageType'].replace(np.nan, 'No Garage')
house['GarageFinish'] = house['GarageFinish'].replace(np.nan, 'No Garage')
house['GarageCond'] = house['GarageCond'].replace(np.nan, 'No Garage')
house['GarageQual'] = house['GarageQual'].replace(np.nan, 'No Garage')
house['BsmtExposure'] = house['BsmtExposure'].replace(np.nan, 'No Basement')
house['BsmtFinType2'] = house['BsmtFinType2'].replace(np.nan, 'No Basement')
house['BsmtFinType1'] = house['BsmtFinType1'].replace(np.nan, 'No Basement')
house['BsmtQual'] = house['BsmtQual'].replace(np.nan, 'No Basement')
house['BsmtCond'] = house['BsmtCond'].replace(np.nan, 'No Basement') 
```

### Correlation matrix

```python
house_corr=house[numlist].corr()
```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(house_corr, annot = True, cmap="coolwarm")
plt.show()
```

```python
n = 100
top_pairs = house_corr.unstack().sort_values(ascending=False)[:n*2]
unique_pairs = []
for pair in top_pairs.index:
    if pair[0] != pair[1] and (pair[1], pair[0]) not in unique_pairs:
        unique_pairs.append(pair)
top_pairs_df = pd.DataFrame(columns=['feature_1', 'feature_2', 'corr_coef'])
for i, pair in enumerate(unique_pairs[:n]):
    top_pairs_df.loc[i] = [pair[0], pair[1], house_corr.loc[pair[0], pair[1]]]
```

```python
top_pairs_df.head(20)
```

```python
#Removing highly correlated variables to avoid multicolinearity
house=house.drop(['GarageCars','TotRmsAbvGrd','TotalBsmtSF', 'GarageYrBlt'] , axis = 1)
```
